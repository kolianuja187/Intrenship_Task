```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import dataset
ds=pd.read_csv("employee_dataset_for_eda.csv")
ds
```

```python
ds.info()
```

# Categorical Variable Analsis

```python
#Employee Count per Department
sns.countplot(data=ds,x='Department',palette='Set2')
plt.title("Employee Count per Department")
```

Finance Department having more count followed by Marketing and Engineering Department.

```python
# Gender Distribution
sns.countplot(data=ds,x='Gender',palette='coolwarm')
plt.title('Gender Distribution')
```

Female Employees having more count as compared to Male Employees.

```python
#Employee Location Distribution
sns.countplot(data=ds,x='Location',palette='Spectral')
plt.title('Number of Employees by Location')
plt.xticks(rotation=45)
```

Seattle Location having more count of Employees as compared to other Location.

```python
# Age Distribution
sns.histplot(ds['Age'],bins=10,kde=True)
plt.title('Age Distribution')
```

45-50 Aged(experienced) Employees having more count followed by 25-30 Aged(Freshers) Employees.

```python
#Years of Experience Distribution
sns.histplot(ds['Years_of_Experience'],bins=15,kde=True)
plt.title('Years of Experience Distribution')
```

2.5-5.0 Years of Experienced Emoloyees having more count followed by 12.5-15.0 years of Experience

```python
#Monthly Salary Distibution
sns.boxplot(data=ds,y='Monthly_Salary',color='lightblue')
plt.title('Salary Spread')
```

Minimum Salary from 20000 to average Salary 75000 to maximum Salary is 120000.

```python
#Average Salary by Department
sns.barplot(data=ds,x='Department',y='Monthly_Salary',estimator=np.mean,palette='viridis')
plt.title('Average Salary by Department')
```

Fianance Department having more Salary followed by Marketing Department

```python
#Performance Score by Department
sns.boxplot(data=ds,x='Department',y='Performance_Score',palette='pastel')
plt.title('Performance Score by Department')
```

Marketing Department having maximum Performance Score then other Department

```python
#Salary Distribution by Gender
sns.violinplot(data=ds,x='Gender',y='Monthly_Salary',palette='muted')
plt.title('Salary Distribution by Gender')
```

Male Employees having more Salary as compared to Female Employees.

```python
sns.heatmap(ds[['Age','Years_of_Experience','Monthly_Salary',
              'Performance_Score']].corr(),annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
```

```python
#Salary vs Experience by Department Scatter Plot
sns.scatterplot(data=ds,x='Years_of_Experience',y='Monthly_Salary',hue='Department')
plt.title('Salary vs. Experience by Department')
```

```python
HR Department having maximum Salary and more Year of Experience
```
